```python
# 练习2：线性函数的梯度
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
x = 2.0
y_true = 5.0

y_pred = w * x + b
L = (y_pred - y_true) ** 2
L.backward()

print(f"w.grad = {w.grad}")
print(f"b.grad = {b.grad}")

print(f"\n手动验证")
print(f"∂L/∂w = 2*(2-5)*2 = {2*(2-5)*2}")
print(f"∂L/∂b = 2*(2-5) = {2*(2-5)}")
```

```python
# 第7格：禁用梯度跟踪
x = torch.tensor(2.0, requires_grad=True)

# 情况1：需要梯度（训练时）
y1 = x ** 2
print(f"训练时: y1.requires_grad = {y1.requires_grad}")

# 情况2：不需要梯度（推理/测试时）
with torch.no_grad():
    y2 = x ** 2
    print(f"推理时: y2.requires_grad = {y2.requires_grad}")

# 或者使用 detach()
y3 = (x ** 2).detach()
print(f"detach后: y3.requires_grad = {y3.requires_grad}")

print(f"\ny1 = {y1}")
print(f"y2 = {y2}")
print(f"y3 = {y3}")   
```

```python
# 第6格：梯度清零(重要！)
x = torch.tensor(2.0, requires_grad=True)

# 第一次计算
y1 = x ** 2
y1.backward()
print(f"第一次求导后: x.grad = {x.grad}")

# 第二次计算（不清零会怎样？）
y2 = x ** 3
y2.backward()
print(f"第二次求导后（未清零）：x.grad = {x.grad}")

# 正确做法：先清零
x.grad.zero_()

y3 = x ** 3
y3.backward()
print(f"第二次求导后（已清零）：x.grad = {x.grad}")
```

```python
# 第5格：向量梯度(重要！)
# 向量输入：y = sum(x^2)，即每个元素平方后求和
x = torch.tensor([1.0, 2.0, 3.0, 4.0], requires_grad=True)
print(f"x = {x}")

# 计算 y = x1^2 + x2^2 + x3^2 + x4^2
y = torch.sum(x ** 2)
print(f"y = sum(x^2) = {y}")

# 求导
y.backward()

print(f"\nx.grad = {x.grad}")
# 应该是 [2, 4, 6, 8]，即 2*x

# 手动验证
print(f"手动: 2*x = {2 *x}")
```

```python
# 第4格：多变量梯度
# 多变量函数：z = x^2 + y^3
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

z = x**2 + y**3

print(f"x = {x}, y = {y}")
print(f"z = x^2 + y^3 = {z}")

# 求导：计算 ∂z/∂x 和 ∂z/∂y
z.backward()

print(f"\n梯度: ")
print(f"∂z/∂x = {x.grad}")
print(f"∂z/∂y = {y.grad}")

# 手动验证
print(f"\n手动验证: ")
print(f"∂z/∂x = 2x = {2 * x.item()}")
print(f"∂z/∂y = 3y^2 = {3 * y.item()**2}")
```

```python
# 第3格：更复杂的函数
import torch

# 定义：y = 3x^2 + 2x + 1
x = torch.tensor(2.0, requires_grad=True)
y = 3 * x**2 + 2 * x + 1

print(f"x = {x}")
print(f"y = 3x^2 + 2x +1 = {y}")

# 求导：dy/dx = 6x + 2 = 6*2 + 2 = 14
y.backward()
print(f"x.grad = {x.grad}")

# 手动验证
manual_grad = 6 * x.item() + 2
print(f"手动计算: 6x + 2 = {manual_grad}")
```

```python
# 第2格：简单的梯度计算
import torch

# 定义：y = x^2
x = torch.tensor(3.0, requires_grad=True)
y = x**2

print(f"x = {x}")
print(f"y = x^2 = {y}")
print(f"y.requires_grad: {y.requires_grad}")

#自动求导：计算dy/dx
y.backward()

print(f"\n求导后: ")
print(f"x.grad = {x.grad}")
print(f"手动计算: 2*x = {2 * x.item()}")
```
